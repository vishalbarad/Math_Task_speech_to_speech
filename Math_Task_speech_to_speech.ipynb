```python
import pytesseract
import cv2
import speech_recognition as sr
from sympy import symbols, solve, Eq
from gtts import gTTS
from IPython.display import Audio
import time
import pytesseract
```

```python
text = pytesseract.image_to_string('Screenshot 2021-04-27 at 7.50.15.jpg')
text = text.rstrip()
text = text.lstrip()
print(text)
```

```python
eq = list()
for i in range(len(text)):
    if(text[i].isnumeric()):
        eq.append(text[i])
        i = i+1
        if(i>=len(text)):
            break
        if(text[i].isalpha()):
            eq.append("*")
        i = i-1
    else:
        eq.append(text[i])    
```

```python
expr1 = list()
k=0
for i in eq:
    k +=1;
    if i is '=':
        break
    else:
        expr1.append(i)
```

```python
expr2 = eq[k:]
```

```python
expr3 = list()
for j in expr2:
    if j=='+':
        expr3.append("-")
    elif j=='-':
        expr3.append("+")
    elif j=='*':
        expr3.append("//")
    elif j=='/':
        expr3.append("*")
    else:
        expr3.append("-")
        expr3.append(j)
```

```python
real_expr1 = ""
real_expr1 = real_expr1.join(expr1)
real_expr2 = ""
real_expr2 = real_expr2.join(expr3)
```

```python
expr = real_expr1+real_expr2
```

```python
ass = list()
for one in range(97,123):
    ass.append(chr(one))

# using list comprehension
listToStr = ','.join(map(str, ass))
listToStr = "'"+listToStr+"'"
```

```python
a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p,q,r,s,t,u,v,w,x,y,z = symbols(listToStr,real=True)
sol = solve(expr)
sol[0]
```

```python
tts = gTTS("done. The answer of "+text+" is "+str(sol[0]))
tts.save("1.wav")
sound_file = "1.wav"
Audio(sound_file,autoplay=True)
```

```python
r = sr.Recognizer()
with sr.Microphone() as source:
    r.adjust_for_ambient_noise(source) 
    print("Say something!")
    audio = r.listen(source)
    try:
        done_speech = r.recognize_google(audio)
        if done_speech.lower() == "dun dun dun":
            message = "User verified successfully"
        elif done_speech.lower() == "done done done":
            message = "User verification successfulu"
        else:
            message = "Could not verify user!"        
    except sr.UnknownValueError:
        message = "Could not understand audio"
    except sr.RequestError as e:
        message = "Could not request results from Google Speech Recognition service;"+e
tts = gTTS(message)
tts.save("1.wav")
sound_file = "1.wav"
Audio(sound_file,autoplay=True)
```
